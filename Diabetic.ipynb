From the above graph Skinthickness, Insulin and BMI has 0 value present in the graph which is not correct.

In the above dataset actually missing values are not present but they refered as 0 which will be seen the above plotted graph.

So missing value is given as the value 0 which is not good for authenticty of dataset. Hence we first replace the 0 value with NaN then we start imputation process.

The columns which 0 value is present are `Glucose`,`BloodPressure`,`Skinthickness`,`Insulin`,`BMI`. 

```python
df['Glucose'] = df['Glucose'].replace(0,np.NaN)
df['BloodPressure'] = df['BloodPressure'].replace(0,np.NaN)
df['SkinThickness'] = df['SkinThickness'].replace(0,np.NaN)
df['Insulin'] = df['Insulin'].replace(0,np.NaN)
df['BMI'] = df['BMI'].replace(0,np.NaN)
```

```python
df.head()
```

```python
# Now check the missing values
df.isna().sum()
```

```python
# Now we will replace missing values with meadian
df['Glucose'].fillna(df['Glucose'].median(),inplace=True)
df['BloodPressure'].fillna(df['BloodPressure'].median(),inplace=True)
df['SkinThickness'].fillna(df['SkinThickness'].median(),inplace=True)
df['Insulin'].fillna(df['Insulin'].median(),inplace=True)
df['BMI'].fillna(df['BMI'].median(),inplace=True)
```

## 4. Feature Scaling

```python
# Create input
X=df.drop(['Outcome'],axis=1)
X.head()
```

```python
# Output
y = df.Outcome
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X = pd.DataFrame((scaler.fit_transform(X)),columns=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age'])
```

```python
X.head()
```

## 5. Split dataset into taining and testing set

```python
# Import train_test_split from sklearn
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
# Check the shape of traing and test data
X_train.shape,X_test.shape
```

## 6. Model Training

```python
# Import the algorithm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

#### Random Forest 

```python
rfc = RandomForestClassifier()
```

```python
param_grid = { 
    'n_estimators': [50,100,200,300],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}
```

```python
# Hyperparameter tuning
from sklearn.model_selection import GridSearchCV
cv_rfc = GridSearchCV(estimator=rfc,param_grid=param_grid,cv=5)
cv_rfc.fit(X_train,y_train)
```

```python
# Now check the best parametes
cv_rfc.best_params_
```

```python
cv_rfc.best_estimator_
```

```python
rfc1 = RandomForestClassifier(max_depth=8,max_features='auto',n_estimators=100,criterion='gini')
```

```python
rfc1.fit(X_train,y_train)
```

```python
rfc1.score(X_train,y_train)*100
```

```python
y_pred_rfc1 = rfc1.predict(X_test)
```

```python
# Import the performance  matrics
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,classification_report
```

```python
# Now check the accuracy of the model
print("Accuracy of the random forest :",accuracy_score(y_test,y_pred_rfc1))
```

```python
# Confusion_matrix
cm = confusion_matrix(y_test,y_pred_rfc1)
cm
```

```python
sns.heatmap(data=cm,annot=True,xticklabels=['Non-diabetic','Diabetic'],yticklabels=['Non-diabetic','Diabetic'])
plt.title('Random forest Confusion matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
# print classification report
print(classification_report(y_test,y_pred_rfc1))
```

```python
# Check for overfitting and underfitting
print('The training score :',rfc1.score(X_train,y_train)*100)

print('The testing score :',rfc1.score(X_test,y_test)*100)
```

The training datset model is overfitted

```python
# cross-validation
from sklearn.model_selection import cross_val_score
```

```python
score = cross_val_score(estimator=rfc1,X=X,y=y,cv=5)
score
```
